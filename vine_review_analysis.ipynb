```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
helpful_votes = vine_df.loc[(vine_df["total_votes"] >= 20)]
helpful_votes
```

```python
helpful_percentage = helpful_votes.loc[(helpful_votes["helpful_votes"] / helpful_votes["total_votes"] >= .5)]
helpful_percentage
```

```python
vine_paid = helpful_percentage.loc[(helpful_percentage["vine"] == 'Y')]
vine_paid.head()
```

```python
vine_unpaid = helpful_percentage.loc[(helpful_percentage["vine"] == 'N')]
vine_unpaid.head()
```

```python
# Paid Total
vine_paid.count()
```

```python
# Vine Paid Total 5-Star Ratings
vine_paid.loc[(vine_paid["star_rating"] == 5.0)].count()
```

```python
# Percentage of Paid 5-Star Reviews
vine_paid.loc[(vine_paid["star_rating"] == 5.0)].count()/vine_paid.count()*100
```

```python
# Unpaid Total
vine_unpaid.count()
```

```python
# Vine Unpaid Total 5-Star Ratings
vine_unpaid.loc[(vine_unpaid["star_rating"] == 5.0)].count()
```

```python
# Percentage of Unpaid 5-Star Reviews
vine_unpaid.loc[(vine_unpaid["star_rating"] == 5.0)].count()/vine_unpaid.count()*100
```
